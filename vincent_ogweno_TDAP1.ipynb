```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
P = np.linspace(0,1,200)
T = np.linspace(0,200,200)
E = np.linspace(-1,1,200)
E1 = np.linspace(0,2,200)

S_P = -(P)*np.log(P)
g_T = np.exp((-1)/T)
b_3_E = 1 /(np.exp(3*(E+1))-1)
b_30_E = 1 /(np.exp(30*(E+1))-1)
f_10_E1 = 1 /(np.exp(10*(E1-1))+1)
f_100_E1 = 1 /(np.exp(100*(E1-1))+1)

fig = plt.figure(1,figsize=(12.2,8.5))
plt.subplot(2,2,1)
plt.plot(P,S_P)
plt.xlabel("P")
plt.ylabel("S(P)")
plt.legend(["S(P)"])
plt.title("Graph of P Vs S(P)")

plt.subplot(2,2,2)
plt.plot(T,g_T)
plt.xlabel("T")
plt.ylabel("g(T)")
plt.legend(["g(T)"])
plt.title("Graph of T Vs g(T)")

plt.subplot(2,2,3)
plt.plot(E,b_3_E,'r')
plt.plot(E[::],b_30_E[::],'g')
plt.xlabel("E")
plt.ylabel("b3(E)")
plt.legend(["b3(E)","b30(E)"])
plt.title("Graph of E Vs b(E)")

plt.subplot(2,2,4)
plt.plot(E1,f_10_E1,'r')
plt.plot(E1[::],f_100_E1[::],'g')
plt.xlabel("E1")
plt.ylabel("f(E1)")
plt.legend(["f_10(E1)","f_100(E1)"])
plt.title("Graph of E Vs f(E)")
plt.show()

```
